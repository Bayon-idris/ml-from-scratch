```python
# Logistic Regression from Scratch

## Objective

Implement logistic regression from scratch using NumPy for binary classification.

## Topics

- Data preprocessing
- Parameter initialization
- Forward propagation
- Binary cross-entropy loss
- Gradient computation
- Gradient descent
- Prediction
- Model evaluation
## Architecture

Input Features
      |
      v
Linear Layer (W,b)
      |
      v
Sigmoid
      |
      v
Probability
      |
      v
Prediction
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from tqdm import tqdm
from sklearn.decomposition import PCA
```

```python
import sys, os
sys.path.append(os.path.abspath(".."))
from utilities.utilities import *
```

```python
def initialisation (X):
    W = np.zeros((X.shape[1], 1))
    b = 0
    return (W, b)
```

```python
def model(X, W, b):
    Z = X.dot(W) + b
    A = 1 / (1 + np.exp(-Z))
    return A
```

```python
def logloss(A, y):
    #Good practice to avoid calculating logarithms at 0
    epsilon = 1e-15
    return 1 / len(y) * np.sum(-y * np.log(A+epsilon) - (1 - y) * np.log(1 - A+epsilon))
```

```python
def compute_gradients(A,X,y):
    dW = 1 / len(y) * np.dot(X.T, A - y)
    db = 1 / len(y) * np.sum(A - y)
    return (dW, db)
```

```python
def update_parameters(W, b, dw, db, lr):
    W = W - lr * dw
    b = b - lr * db
    return W, b
```

```python
def predict(X,W,b):
    A = model(X, W, b)
    # print(A)
    return A>0.5
```

```python
def preprocess(X):
    X = X.reshape(X.shape[0], -1)
    X = X/255.0
    return X
```

```python
def logistic_regression(X_train, y_train, X_test, y_test, lr=0.01, nb_iter=1000):
    W, b = initialisation(X_train)
    train_losses, test_losses = [], []
    train_accs, test_accs = [], []

    for i in tqdm(range(nb_iter)):
        # Forward train
        A_train = model(X_train, W, b)
        train_losses.append(logloss(A_train, y_train))
        y_pred_train = (A_train > 0.5).astype(int)
        train_accs.append(accuracy_score(y_train, y_pred_train))

        # Forward test
        A_test = model(X_test, W, b)
        y_pred = predict(X_test, W, b)
        test_losses.append(logloss(A_test, y_test))
        y_pred_test = (A_test > 0.5).astype(int)
        test_accs.append(accuracy_score(y_test, y_pred_test))

        # Gradients + update
        dW, db = compute_gradients(A_train, X_train, y_train)
        W, b = update_parameters(W, b, dW, db, lr)

    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(test_losses, label="Test Loss")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(train_accs, label="Train Accuracy")
    plt.plot(test_accs, label="Test Accuracy")
    plt.xlabel("Iterations")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()

    return W, b
```

```python
X_train, y_train, X_test, y_test = load_data()
X_train = preprocess(X_train)
X_test = preprocess(X_test)
```

```python
def plot_decision_boundary(X_train, y_train, X_test, y_test, W, b):
    
    # 2D reduction if necessary
    if X_train.shape[1] > 2:
        pca = PCA(n_components=2)
        X_train_2d = pca.fit_transform(X_train)
        X_test_2d = pca.transform(X_test)
        W_2d = pca.components_.dot(W)  # projection des poids
        b_2d = b
    else:
        X_train_2d, X_test_2d = X_train, X_test
        W_2d, b_2d = W, b

    x_min, x_max = X_train_2d[:, 0].min() - 1, X_train_2d[:, 0].max() + 1
    y_min, y_max = X_train_2d[:, 1].min() - 1, X_train_2d[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))

    grid = np.c_[xx.ravel(), yy.ravel()]
    Z = 1 / (1 + np.exp(-(np.dot(grid, W_2d) + b_2d)))
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, levels=[0, 0.5, 1], alpha=0.3, colors=["#FFAAAA", "#AAAAFF"])
    plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y_train, cmap=plt.cm.RdBu, edgecolors="k", label="Train")
    plt.scatter(X_test_2d[:, 0], X_test_2d[:, 1], c=y_test, cmap=plt.cm.RdBu, marker="x", label="Test")
    plt.legend()
    plt.title("Decision Boundary for LR")
    plt.show()
```

```python
W, b = logistic_regression(X_train, y_train, X_test, y_test, lr=0.001, nb_iter=1000)
y_pred = predict(X_test, W, b)
print("Test Accuracy :", accuracy_score(y_test, y_pred))
plot_decision_boundary(X_train, y_train, X_test, y_test, W, b)
```

```python
In this notebook, we implemented Logistic Regression for binary image classification. Despite its simplicity, this model provides a solid introduction to supervised learning and gradient-based optimization.

### Limitations of Logistic Regression

Logistic Regression is a **linear model**. It attempts to separate classes using a single decision boundary and therefore struggles to capture complex patterns in image data.

For image classification tasks such as distinguishing cats from dogs, its limitations quickly become apparent:

- It cannot automatically detect visual features such as edges, shapes, or textures.
- It assumes a linear relationship between the input pixels and the output class.
- Its performance is limited when dealing with complex, non-linear data.

### Moving Beyond Linear Models

To overcome these limitations, we need models capable of learning **non-linear relationships**. Neural Networks introduce hidden layers that can learn increasingly complex representations of the input data.
Instead of working directly with raw pixels, these networks can learn useful intermediate features that improve classification performance.

➡️ Next step: **03_neural_network_one_hidden_layer.ipynb**

In the next notebook, we will build our first Neural Network with a hidden layer and explore how it can outperform Logistic Regression on image classification tasks.
```
